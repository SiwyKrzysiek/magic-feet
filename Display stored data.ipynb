```python
import pandas as pd
import matplotlib.pyplot as plt
import redis
import json
```

Connect to Redis

```python
store = redis.Redis()
```

Read data

```python
raw = store.lrange('personData2', 0, -1) # -1 is the last element

data = [json.loads(d.decode()) for d in raw]
data
```
